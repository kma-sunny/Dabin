```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

```python
train=pd.read_csv("train.csv",parse_dates=["datetime"])
```

```python
train.info()
```

```python
train.head()
```

```python
sns.distplot(train["count"])
```

```python
train.temp.describe()
```

```python
train['year']=train["datetime"].dt.year
train['month']=train["datetime"].dt.month
train['day']=train["datetime"].dt.day
train['hour']=train["datetime"].dt.hour
train['minute']=train["datetime"].dt.minute
train['second']=train["datetime"].dt.second
train.head()
```

```python
# sns.barplot(data=train,x="year",y="count")
sns.barplot(data=train,x="month",y="count")
```

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2)
sns.barplot(data=train,x="year",y="count",ax=ax1)
sns.barplot(data=train,x="month",y="count",ax=ax2)
sns.barplot(data=train,x="day",y="count",ax=ax3)
sns.barplot(data=train,x="hour",y="count",ax=ax4)

ax1.set(ylabel='count',title='연도별대여량')
```

```python
fig, axes = plt.subplots(2,2)
fig.set_size_inches(12,10)
sns.boxplot(data=train, y='count', x='season',ax=axes[0][0])
sns.boxplot(data=train, y='count', x='workingday',ax=axes[0][1])
sns.boxplot(data=train, y='count', x='season',ax=axes[1][0])
sns.boxplot(data=train, y='count', x='season',ax=axes[1][1])
```

```python
train["dayofweek"]=train["datetime"].dt.dayofweek #요일 0-6
train['dayofweek'].value_counts()
```

```python
fig, ax= plt.subplots(2)
fig.set_size_inches(110,8)
sns.pointplot(data=train, x="hour",y="count",ax=ax[0])
sns.pointplot(data=train, x="hour",y="count",hue="workingday",ax=ax[1])
```

```python
corrMatt=train[['temp','atemp','casual', 'registered','humidity','windspeed','count']]
train.columns
```

```python
corrMatt=corrMatt.corr()
```

```python
corrMatt
```

```python
mask=np.array(corrMatt)
mask[np.tril_indices_from(mask)]=False
mask
```

```python
sns.heatmap(corrMatt, mask=mask,vmax=.8, square=True,annot=True)
fig.set_size_inches(20,10)
```

```python
fig,ax=plt.subplots(3)
fig.set_size_inches(12,6)

sns.regplot(y='temp',x='count',data=train,ax=ax[0])
sns.regplot(y='windspeed',x='count',data=train,ax=ax[1])
sns.regplot(x='humidity',y='count',data=train,ax=ax[2])
```

```python
def ym(datetime):
    return "{0}-{1}".format(datetime.year,datetime.month)

train["ym"]=train["datetime"].apply(ym)

fig, ax3 = plt.subplots(1)
fig.set_size_inches(18, 4)

sns.barplot(data=train, x='ym',y='count')
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
fig.set_size_inches(18, 4)

sns.barplot(data=train, x="year", y="count", ax=ax1)
sns.barplot(data=train, x="month", y="count", ax=ax2)
```
