# WEATHER

```python
temp_data_original=pd.read_csv("격자_월_일별_온도_풍속_체감온도.csv")
```

```python
temp_data_original.head(2)
```

```python
temp_data_06=temp_data_original[temp_data_original["month"]==6]
temp_data_07=temp_data_original[temp_data_original["month"]==7]
temp_data_08=temp_data_original[temp_data_original["month"]==8]
```

```python
temp_data_07=temp_data_07[temp_data_07["day"]!=31]
temp_data_08=temp_data_08[temp_data_08["day"]!=31]
```

```python
temp_data_06_guro=temp_data_06[temp_data_06["grid"]=="57_125"]
temp_data_07_guro=temp_data_07[temp_data_07["grid"]=="57_125"]
temp_data_08_guro=temp_data_08[temp_data_08["grid"]=="57_125"]
```

```python
def make_round(x):
    return round(x,1)
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
temp_data_06_guro["temp_avg"]=temp_data_06_guro["temp_avg"].map(make_round)
temp_data_07_guro["temp_avg"]=temp_data_07_guro["temp_avg"].map(make_round)
temp_data_08_guro["temp_avg"]=temp_data_08_guro["temp_avg"].map(make_round)
```

```python
attr=temp_data_07["day"].unique()

v1 =temp_data_06_guro["temp_avg"]
v2 =temp_data_07_guro["temp_avg"]
v3 =temp_data_08_guro["temp_avg"]

line =Line("구로구 평균기온의 변화")
line.add("6월", attr, v1, mark_point=["max", "min"])
line.add("7월", attr, v2, mark_point=["max", "min"])
line.add("8월", attr, v3, mark_point=["max", "min"], yaxis_formatter="°C")
line
```
